## MatplotLib Basic Bar Graph 

**Benefits**
- intuitive 
- was able to somewhat mimic prism style for my poster 

**Drawbacks** 
- making any single graph in python look good is harder than using a GUI like prism , but this is probably as easy as it can be 


```python
#for PEAK diagnoses effect bar graph 

#import necessary modules 
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 

# Create lists for the plot
diagnoses = ['HC', 'MDD']
HC=.1539
MDD=.5175
HC_err=.09257
MDD_err=.08375

#determine positions on the bar graph 
x_pos = np.arange(len(diagnoses))
effects = [HC,MDD]
error = [HC_err,MDD_err]

#customize the bar graph with plt subplots 
fig, ax = plt.subplots()
ax.bar(x_pos, effects, yerr=error, align='center', color = ['orange','steelblue'], ecolor='black', capsize=10)
ax.set_ylabel('Peak', fontsize='14')
plt.rcParams["figure.figsize"]=[5,5]
ax.set_xticks(x_pos)
ax.set_xticklabels(diagnoses, fontsize='14')
ax.set_title('Left Mid Occipital Peak by DX', fontsize='18')
ax.yaxis.grid(False)
plt.tick_params(labelsize=13)

# Save the figure and show

plt.autoscale()
plt.tight_layout()
plt.savefig('/home/sepeforrestl2/EmoEvalGraphs/LMidOcc_Diagnoses_Bar_KETPLC.png',format='png', dpi=1000)
```

**Benefits**

- Seaborn makes it easy to do all of this in one graph while other datavisualization tools might require multiple graphs to do this

```python
#import necessary packages 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib as plt 

#read in and look at dataset  
data = pd.read_csv('yearly_deaths_by_clinic.csv')
print(data.head())

#create plot and customize colors 
ax=sns.violinplot("clinic", "deaths", data=data,
               palette=["steelblue", "orange"])

#set titles and labels 
ax.set_title('Births and Deaths', fontsize='18')
ax.set_ylabel('Deaths', fontsize='14')
ax.set_xlabel('Clinic', fontsize='14')

#further customize graph by offsetting x axis and setting styles and context 
sns.despine(offset=.5, trim=True)
sns.set_style("ticks", {"axes.facecolor": "1"})
sns.set_context("paper", font_scale=1.5)
```

### Interactive features- HoverTool

```python
# output_notebook to display the graph inline
output_notebook()

# Import modules
from bokeh.plotting import ColumnDataSource, figure
from bokeh.models import HoverTool
import pandas as pd

# Read in dataset
df = pd.read_csv('yearly_deaths_by_clinic.csv')

# Create a ColumnDataSource from df: source
source = ColumnDataSource(df)

# Create a figure with the "box_select" tool: p
p = figure(x_axis_label='year', y_axis_label='deaths', tools='box_select')

# Add circle glyphs to the figure p with the selected and non-selected properties
p.circle(selection_color='red', nonselection_alpha=.1, x='year', y='deaths', source=source)

# Create a HoverTool: hover
hover = HoverTool(tooltips=None, mode='vline')

# Add the hover tool to the figure p
p.add_tools(hover)

# Specify the name of the output file and show the result
show(p)
```

*Connect slider to figure*

```python
# output_notebook to display the graph inline
output_notebook()

# Import modules
import numpy as np

from bokeh.layouts import row, column
from bokeh.models import CustomJS, Slider
from bokeh.plotting import figure, output_file, show, ColumnDataSource

# Define variables
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Define source
source = ColumnDataSource(data=dict(x=x, y=y))

# Originate graph
plot = figure(y_range=(-10, 10), plot_width=400, plot_height=400)

plot.line('x', 'y', source=source, line_width=3, line_alpha=0.6)

callback = CustomJS(args=dict(source=source), code="""
    var data = source.data;
    var A = amp.value;
    var k = freq.value;
    var phi = phase.value;
    var B = offset.value;
    var x = data['x']
    var y = data['y']
    for (var i = 0; i < x.length; i++) {
        y[i] = B + A*Math.sin(k*x[i]+phi);
    }
    source.change.emit();
""")

# Define slider fuctions
amp_slider = Slider(start=0.1, end=10, value=1, step=.1,
                    title="Amplitude", callback=callback)
callback.args["amp"] = amp_slider

freq_slider = Slider(start=0.1, end=10, value=1, step=.1,
                     title="Frequency", callback=callback)
callback.args["freq"] = freq_slider

phase_slider = Slider(start=0, end=6.4, value=0, step=.1,
                      title="Phase", callback=callback)
callback.args["phase"] = phase_slider

offset_slider = Slider(start=-5, end=5, value=0, step=.1,
                       title="Offset", callback=callback)
callback.args["offset"] = offset_slider

layout = row(
    plot,
    column(amp_slider, freq_slider, phase_slider, offset_slider),
)

output_file("slider.html", title="slider.py example")

show(layout)
```

## Choropleth

```python
# Import modules
import altair as alt
from vega_datasets import data

# Define data
states = alt.topo_feature(data.us_10m.url, 'states')
source = data.population_engineers_hurricanes.url
variable_list = ['population', 'engineers']

# Define parameters
alt.Chart(states).mark_geoshape().encode(
    alt.Color(alt.repeat('row'), type='quantitative')
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(source, 'id', variable_list)
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
).repeat(
    row=variable_list
).resolve_scale(
    color='independent'
)
```
